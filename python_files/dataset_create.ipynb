```python
## creating a function that will take the path of the file and create/merge the data


ids = []
texts = []
labels = []

def read_json_df(path):
  with open(path, "r") as file:
    for line in file:
      data = json.loads(line)

      # print(data)
      # break

      id_data = data.get("id")
      text_data = data.get("text")
      label_data = data.get("label")


      ## merging all available data
      ids.append(id_data)
      texts.append(text_data)
      labels.append(label_data)

  ## creating dataframe
  data = pd.DataFrame({"id" : ids,
                      "text" : texts,
                      "label" : labels})
  return data
```

```python
## using function to create the final data

final_data = pd.DataFrame()

paths = ["Data/GBRAIN_TRAIN_10k.jsonl",
         "Data/GLTR_gpt2xl_TRAIN_k40_temp07_4k.jsonl"]




for path in paths:
  data = read_json_df(path)
  final_data = pd.concat([final_data, data])


# Reset index after concatenation
final_data.reset_index(drop=True, inplace=True)

# Display final_data
# print(final_data.label.value_counts())
```
