## Reading and Loading the data with necessary packages


```python
## importing packages

import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

```

```python
## loading the data

detectllm = pd.read_csv("Data/Dataset/DetectLLM.csv")
detectllm.shape
```

```python
detectllm.head()
```

## Confusion Matrix and Result Generation
---------------------------------

```python
from sklearn.metrics import confusion_matrix

# Define a function to generate the Generated_Label based on LLM_generated_perc(from_calculation)
def generate_generated_label(row):
    if row['LLM_generated_perc(from_calcualtion)'] > 50:
        return 'LLMs'
    else:
        return 'Human'

# Apply the function to create the Generated_Label column
detectllm['Generated_Label'] = detectllm.apply(generate_generated_label, axis=1)

# Create confusion matrix
conf_matrix = confusion_matrix(detectllm['Label'], detectllm['Generated_Label'])

# Display confusion matrix
print("Confusion Matrix:")
print(conf_matrix)
```

```python
from sklearn.metrics import accuracy_score, f1_score, classification_report

# Calculate accuracy score
accuracy = accuracy_score(detectllm['Label'], detectllm['Generated_Label'])

# Calculate F1 score
f1 = f1_score(detectllm['Label'], detectllm['Generated_Label'], average='weighted')

# Generate classification report
report = classification_report(detectllm['Label'], detectllm['Generated_Label'])

# Display results
print("Accuracy Score:", accuracy)
print("F1 Score:", f1)
print("Classification Report:\n", report)
```

### Plotting the result
-----------------------------

```python
# Convert classification report to dictionary
report_dict = classification_report(detectllm['Label'], detectllm['Generated_Label'], output_dict=True)

# Convert dictionary to DataFrame
report_df = pd.DataFrame(report_dict).transpose()

# Plot heatmap
plt.figure(figsize=(7,4))
# sns.heatmap(report_df.iloc[:-1, :-1], annot=True, cmap='viridis', fmt='.2f', linewidths=.5)
# sns.heatmap(report_df.iloc[:-1, :-1], annot=True, cmap='magma', fmt='.2f', linewidths=.5)
sns.heatmap(report_df.iloc[:-1, :-1], annot=True, cmap='inferno', fmt='.2f', linewidths=.5)
# sns.heatmap(report_df.iloc[:-1, :-1], annot=True, cmap='plasma', fmt='.2f', linewidths=.5)
plt.title('Classification Report - Perc-DETECTLLM')
plt.xlabel('Metrics')
plt.ylabel('Classes')
plt.show()
```
