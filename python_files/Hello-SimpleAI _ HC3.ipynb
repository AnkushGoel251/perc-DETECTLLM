```python
import pandas as pd
import numpy as np
import json
```

```python
## reading the data

questions = []
chatgpt_answers = []
human_answers = []
source = []

with open("/content/drive/MyDrive/BTP 8th SEM/Data/Hello-SimpleAI HC3.jsonl", "r") as file:
  for line in file:
    data = json.loads(line)

    '''
    Collecting datapoints for all individual data
    '''
    question_data = data.get('question')
    chatgpt_answer_data = data.get('chatgpt_answers')
    human_answer_data = data.get('human_answers')
    source_data = data.get("source")

    '''
    Creating final dataset
    '''

    questions.append(question_data)
    chatgpt_answers.append(chatgpt_answer_data)
    human_answers.append(human_answer_data)
    source.append(source_data)


## creating the dataframe
hello_simpleAI_HC3_data = pd.DataFrame({"question" : questions,
                                        "chatgpt_answer" : chatgpt_answers,
                                        "human_answer" : human_answers,
                                        "source" : source})
hello_simpleAI_HC3_data.shape
```
