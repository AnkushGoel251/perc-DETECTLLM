```python
## importing packages

import pandas as pd
import numpy as np
```

```python
## loading the data

## loading the data

detectllm = pd.read_csv("/content/drive/MyDrive/BTP 8th SEM/Data/Dataset/DetectLLM.csv")
detectllm.shape
```

```python
detectllm.head()
```

```python
llm_generated = detectllm[detectllm['Label'] == 'LLMs']
human_generated = detectllm[detectllm['Label'] == 'Human']
```

## Data Information
-----------------

### 1. All

```python
# Remove missing values from 'Improved_Texts' column
detectllm_cleaned = detectllm.dropna(subset=['Improved_Texts'])

# Calculate Average Document Length
average_doc_length = detectllm_cleaned['Improved_Texts'].apply(lambda x: len(str(x).split())).mean()

# Calculate Average Sentence Length
average_sentence_length = detectllm_cleaned['Improved_Texts'].apply(lambda x: len(str(x).split()) / len(str(x).split('.'))).mean()

# Calculate Average Number of Sentences per Document
average_sentences_per_doc = detectllm_cleaned['Improved_Texts'].apply(lambda x: len(str(x).split('.'))).mean()

print("Average Document Length:", round(average_doc_length, 2))
print("Average Sentence Length:", round(average_sentence_length, 2))
print("Average Number of Sentences per Document:", round(average_sentences_per_doc, 2))

```
